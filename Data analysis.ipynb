```python
import os

train_dir = "C:\\Users\\Windows 10\\Videos\\psd\\Dataset\\Splitted\\train"
val_dir = "C:\\Users\\Windows 10\\Videos\\psd\\Dataset\\Splitted\\val"

Bacterialblight_train_path = train_dir + "/Bacterialblight"
Blast_train_path = train_dir + "/Blast"
Brownspot_train_path = train_dir + "/Brownspot"
Tungro_train_path = train_dir + "/Tungro"
Bacterialblight_val_path = val_dir + "/Bacterialblight"
Blast_val_path = val_dir + "/Blast"
Brownspot_val_path = val_dir + "/Brownspot"
Tungro_val_path = val_dir + "/Tungro"

Bacterialblight_len_train = len(os.listdir(Bacterialblight_train_path))
Blast_len_train = len(os.listdir(Blast_train_path))
Brownspot_len_train = len(os.listdir(Brownspot_train_path))
Tungro_len_train = len(os.listdir(Tungro_train_path))
Bacterialblight_len_val = len(os.listdir(Bacterialblight_val_path))
Blast_len_val = len(os.listdir(Blast_val_path))
Brownspot_len_val = len(os.listdir(Brownspot_val_path))
Tungro_len_val = len(os.listdir(Tungro_val_path))

print("Jumlah dataset training : ", Bacterialblight_len_train + Blast_len_train + Brownspot_len_train + Tungro_len_train)
print("Jumlah dataset validasi : ", Bacterialblight_len_val + Blast_len_val + Brownspot_len_val + Tungro_len_val)
print()
print("Jumlah train kelas Bacterialblight ", Bacterialblight_len_train)
print("Jumlah train kelas Blast : ", Blast_len_train)
print("Jumlah train kelas Brownspot ", Brownspot_len_train)
print("Jumlah train kelas Tungro : ", Tungro_len_train)
print("Jumlah validasi kelas Bacterialblight : ", Bacterialblight_len_val)
print("Jumlah validasi kelas Blast : ", Blast_len_val)
print("Jumlah validasi kelas Brownspot : ", Brownspot_len_val)
print("Jumlah validasi kelas Tungro : ", Tungro_len_val)
```

```python
train_Bacterialblight_fnames = os.listdir(Bacterialblight_train_path)
train_Blast_fnames = os.listdir(Blast_train_path)
train_Brownspot_fnames = os.listdir(Brownspot_train_path)
train_Tungro_fnames = os.listdir(Tungro_train_path)

print(train_Bacterialblight_fnames[:10])
print(train_Blast_fnames[:10])
print(train_Brownspot_fnames[:10])
print(train_Tungro_fnames[:10])
```

```python
# import terlebih dahulu library yang diperlukan

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Parameter dari visualisasi dimana terdiri dari 4 kolom
nrows = 4
ncols = 4

# Inisialisasi nilai awal untuk melakukan iterasi
pic_index = 0
```

```python
# Mengatur matplotlib fig, dan ukuran visualisasi ke 4x4 pics

fig = plt.gcf()
fig.set_size_inches(ncols*4, nrows*4)

pic_index += 4

next_Bacterialblight_pix = [os.path.join(Bacterialblight_train_path, fname)
                   for fname in train_Bacterialblight_fnames[pic_index-4:pic_index]
                  ]

next_Blast_pix = [os.path.join(Blast_train_path, fname)
                   for fname in train_Blast_fnames[pic_index-4:pic_index]
                  ]

next_Brownspot_pix = [os.path.join(Brownspot_train_path, fname)
                   for fname in train_Brownspot_fnames[pic_index-4:pic_index]
                  ]

next_Tungro_pix = [os.path.join(Tungro_train_path, fname)
                   for fname in train_Tungro_fnames[pic_index-4:pic_index]
                  ]

for i, img_path in enumerate(next_Bacterialblight_pix+next_Blast_pix+next_Brownspot_pix+next_Tungro_pix):
  # Menyeting subplot
  sp = plt.subplot(nrows, ncols, i + 1)
  # Tidak menampilkan axes (atau gridlines)
  sp.axis("off")

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show() 
```
